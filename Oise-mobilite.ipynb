### On importe les lib python

```python
import json #pour la gestion des données
import folium #pour l'affichage de la carte
```

### On lit le fichier json et on l'affiche

```python
with open("ligne_618.json") as jsonFile:
    jsonObject = json.load(jsonFile)

#print(json.dumps(jsonObject, indent=4)) # pour visualiser le fichier json
```

### on récupère les données pour notre itinéraire et on boucle dessus pour récupérer le tracé de la ligne dans un dictionaire python

```python
iti = jsonObject['Data']['Itineraries'] #la balise qui contiend le tracé des lignes
print("nombre d'itinéraires : {}. variable de type : {}".format(len(iti), type(iti)))


for i in iti:
    geometries = i['Geometries'][0]
    print("Nb de clés : {}".format(len(geometries)))

seg = geometries['Geometry']
print("Nb de ségments : {}. variable de type {}".format(len(seg), type(seg)))

lst_seg = []
for s in seg:
    dict_seg={
        "type": "Feature",
        "geometry": {
            "type":s['type'],
            "coordinates":s['coordinates']
            
        }
    }
    lst_seg.append(dict_seg)
#print(lst_seg)
```

### on récupère les données pour notre itinéraire et on boucle dessus pour récupérer les arrêts dans un dictionaire python

```python
stops = jsonObject['Data']['Stops']
print("nombre d'arrêts : {}. variable de type : {}".format(len(stops), type(stops)))

lst_stops = []
for s in stops:
    #print(type(s), s)
    dict_stop = {
        "geometry": {
            "type":"Point",
            "coordinates": [s['Longitude'], s['Latitude']]
        },
        "type": "Feature",
        "properties": {
            "name":s['Name']
        }        
    }
    lst_stops.append(dict_stop)
#print(lst_stops)
```

### on reconstitu une variable géojson avec les éléments récupérés

```python

line_geojson = {
    "type":"FeatureCollection",
    "features":lst_seg
}
pt_geojson = {
    "type":"FeatureCollection",
    "features":lst_stops
}
```

### on affiche la variable geojson sur une carte

```python
carte = folium.Map(
    location=[49.4009900093076, 1.81146001815776],
    tiles="Stamen Terrain",
    zoom_start=11,
)

folium.GeoJson(line_geojson, name="line").add_to(carte)
folium.GeoJson(pt_geojson, name="stops").add_to(carte)
carte
```

### on exporte le résultat pour une ouverture dans QGis...

```python
with open('segments.json', 'a') as f:
    json.dump(line_geojson, f)
with open('stops.json', 'a') as f:
    json.dump(pt_geojson, f)
```
